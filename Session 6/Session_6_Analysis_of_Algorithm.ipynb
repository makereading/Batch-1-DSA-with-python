### What is Datastructures ?

Data structure is a particular way of **storing** and **organizing** data in a computer so that it can be used efficientlly.

A datastructure is a special format for organizing and storing data. General data structure types include 

1. arrays
2. files
3. linked lists
4. stacks
5. queues
6. trees
7. *graphs* and so on.

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
def f_of_n(n):
  return n**2 + n + 1

X = [1, 2, 3, 4, 5, 6, 7, 10, 15, 23, 34, 35]
Y = []
for x_val in X:
  Y.append(f_of_n(x_val))
print(Y)
```

```python
plt.plot(X, Y, color='#8f3a84')
plt.xlabel("X Values")
plt.ylabel("F of N")
plt.legend()
plt.title("Plotting a random function")
plt.show()
```

```python
def g_of_n(n):
  return 2*n**2
Y_generated_from_g_of_n = []
for x_val in X:
  Y_generated_from_g_of_n.append(g_of_n(x_val))
print(Y_generated_from_g_of_n)
```

```python
plt.plot(X, Y, color='#8f3a84')
plt.plot(X, Y_generated_from_g_of_n, color='blue')
plt.xlabel("X Values")
plt.ylabel("F of N")
plt.legend()
plt.title("Plotting a random function")
plt.show()
```

## Theta - Average Bound

```python
import math

def f_of_n(n):
  return 4*n + 10

def g_of_n(n):
  return 5*n

def c_of_n(n):
  return 4*n


X = [1, 2, 3, 4, 5, 6, 7, 10, 15, 23, 34, 35]
Y = []
for x_val in X:
  Y.append(f_of_n(x_val))

Y_generated_from_g_of_n = []
for x_val in X:
  Y_generated_from_g_of_n.append(g_of_n(x_val))

Y_generated_from_c_of_n = []
for x_val in X:
  Y_generated_from_c_of_n.append(c_of_n(x_val))

plt.plot(X, Y, color='#8f3a84', linewidth=1)
plt.plot(X, Y_generated_from_g_of_n, color='blue', linewidth=1)
plt.plot(X, Y_generated_from_c_of_n, color='green', linewidth=1)
plt.xlabel("X Values")
plt.ylabel("F of N")
plt.legend()
plt.title("Plotting a random function")
plt.show()
```

## Order of growth

image.png

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 7]

import numpy as np
import math
 
n = np.linspace(2,3)
print(n)
plt.plot(n, [0.1]*(len(n)), label="constant")
plt.plot(n, np.log2(n), label="logn")
plt.plot(n, np.sqrt(n), label="sqrt(n)")
plt.plot(n, n, label="n")
plt.plot(n, np.log(n), label="nlogn")
plt.plot(n, np.power(n,2), label="n^2")
plt.plot(n, np.power(2,n), label="2^n")
plt.legend(loc='upper left')
plt.show()
```
