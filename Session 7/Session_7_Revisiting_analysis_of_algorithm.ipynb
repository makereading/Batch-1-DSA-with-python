```python
# Function to plot lines
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.rcParams["figure.figsize"] = (20,10)

def plot_it(lines, x_label= "X Values", y_label = "F of N"):
  for line in lines:
    plt.plot(line[0], line[1], color=line[2], linewidth=1)
  plt.xlabel(x_label)
  plt.ylabel(y_label)
  plt.legend()
  plt.title("Plotting a random function")
  plt.show()
```

Lets define a function/problem/formula. We are going to find how good , how bad, how average it could confirm. 



f(n) = n2 + n + 1

```python
def f_of_n(n):
  return n*n + n + 1

X = [1, 2, 3, 4, 5, 6, 7, 10, 15, 23, 34, 35]
Y = []
for x_val in X:
  # print(x_val, f_of_n(x_val))
  Y.append(f_of_n(x_val))

print(Y)
plot_it([[X, Y, 'blue']])
```

```python
# f(n) = n2 + n + 1
# g(n) = 2*n*n
```

```python
def g_of_n(n):
  return 2*n*n + 2

Y_generated_from_g_of_n = []
for x_val in X:
  Y_generated_from_g_of_n.append(g_of_n(x_val))
print(Y)
print(Y_generated_from_g_of_n)

plot_it([[X, Y, 'blue'], [X, Y_generated_from_g_of_n, 'yellow']])
```

```python
# f(n) = n2 + n + 1
def g_of_n(n):
  return n*n + n + 10

Y_generated_from_g_of_n = []
for x_val in X:
  Y_generated_from_g_of_n.append(g_of_n(x_val))
print(Y)
print(Y_generated_from_g_of_n)

plot_it([[X, Y, 'blue'], [X, Y_generated_from_g_of_n, 'yellow']])
```
